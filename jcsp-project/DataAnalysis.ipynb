```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("TestData/TestData1.csv")
df.head(5)
```

```python
def plot_test_results(df, test_num, ax=None, colour='b', title=""):
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(12,8))

    test_data = df[df['TestNumber'] == test_num]
    test_data.head(5)

    ax.plot(test_data["numProductions"], test_data["TimeMillis"]/1000, color=colour)
    ax.set_title(title)
    ax.set_xlabel("Total Productions")
    ax.set_ylabel("Time [s]")

```

```python
fig, ax = plt.subplots(2, 4, figsize = (20,12))
plot_test_results(df, 1, ax[0, 0], 'b', "Test1\nP: 1, C: 1, B: 1")
plot_test_results(df, 2, ax[0, 1], 'r', "Test2\nP: 10, C: 1, B: 1")
plot_test_results(df, 3, ax[0, 2], 'g', "Test3\nP: 1, C: 10, B: 1")
plot_test_results(df, 5, ax[0, 3], 'orange', "Test4\nP: 100, C: 100, B: 10")

plot_test_results(df, 1, ax[1, 0], 'b')
plot_test_results(df, 2, ax[1, 0], 'r', "Test1 + Test2")

plot_test_results(df, 1, ax[1, 1], 'b')
plot_test_results(df, 3, ax[1, 1], 'g', "Test1 + Test3")

plot_test_results(df, 1, ax[1, 2], 'b')
plot_test_results(df, 5, ax[1, 2], 'orange', "Test1 + Test4")

test_data = df[df['TestNumber'] == 4]
# test_data.head(5)
ax[1, 3].set_ylim(0, 3.2)
ax[1, 3].plot(test_data["numBuffers"], test_data["TimeMillis"]/1000, color='brown')
ax[1, 3].set_title("Test5\nP: 5, C: 5, Productions: 10000")
ax[1, 3].set_xlabel("Total Productions")
ax[1, 3].set_ylabel("Time [s]")

plt.show()
```
